### **Reading Data**
This data is collected from **ISNA** news agency and contains title, link, abstract, paragraphs and tags of news. You can find the crawling script in **crawling.py**.

```python
import pandas as pd

data = pd.read_json('data.json')
data
```

### **Normalization**
In order to do the task, we have to normalize all words in news.

```python
from hazm import *

normalizer = Normalizer()

for i in range(len(data)):
    data.iloc[i, 0] = normalizer.normalize(data.iloc[i, 0])
    data.iloc[i, 2] = normalizer.normalize(data.iloc[i, 2])
    data.iloc[i, 3] = [normalizer.normalize(r) for r in data.iloc[i, 3]]
    data.iloc[i, 4] = [normalizer.normalize(r) for r in data.iloc  [i, 4]]
```

### **Tokenization**
Then we tokenize our text so that we can access to it's words. In order to do that, **first we tokenize the sentences and then, we tokenize the words**.


#### **tokenizing sentences:**

```python
for i in range(len(data)):
    data.iloc[i, 0] = sent_tokenize(data.iloc[i, 0])
    data.iloc[i, 2] = sent_tokenize(data.iloc[i, 2])
    data.iloc[i, 3] = [normalizer.normalize(r) for r in data.iloc[i, 3]]

data
```

#### **tokenizing words:**

```python
for i in range(len(data)):
    data.iloc[i, 0] = [word_tokenize(s) for s in data.iloc[i, 0]]
    data.iloc[i, 2] = [word_tokenize(s) for s in data.iloc[i, 2]]
    data.iloc[i, 3] = [word_tokenize(s) for s in data.iloc[i, 3]]
    data.iloc[i, 4] = [word_tokenize(s) for s in data.iloc[i, 4]]
```

```python
data
```

### **Collecting All Tokens Together**
Now we need to collect all tokens together:

```python
tokens = []
title_tokens = []
abstract_tokens = []
paragraphs_tokens = []
tag_tokens = []
for i in range(len(data)):
    [tokens.append(e) for e in data.iloc[i, 0]]
    [tokens.append(e) for e in data.iloc[i, 2]]
    [tokens.append(e) for e in data.iloc[i, 3]]
    [tokens.append(e) for e in data.iloc[i, 4]]
    [title_tokens.append(e) for e in data.iloc[i, 0]]
    [abstract_tokens.append(e) for e in data.iloc[i, 2]]
    [paragraphs_tokens.append(e) for e in data.iloc[i, 3]]
    [tag_tokens.append(e) for e in data.iloc[i, 4]]

```

```python
tokens
```

```python
def flatten_list(input_list):
    output_list = []
    for item in input_list:
        if isinstance(item, list):
            output_list.extend(flatten_list(item))
        else:
            output_list.append(item)
    return output_list

tokens = flatten_list(tokens)
title_tokens = flatten_list(title_tokens)
abstract_tokens = flatten_list(abstract_tokens)
paragraphs_tokens = flatten_list(paragraphs_tokens)
tag_tokens = flatten_list(tag_tokens)

tokens
```

```python
from nltk import FreqDist

df = {}

for col in ['title_', 'abstract_', 'paragraphs_', 'tag_', '']:
    df[col] = FreqDist(eval(f'{col}tokens')).most_common(10) 
freq_analysis = pd.DataFrame(df) 
```

```python
freq_analysis
```

### **Stop Words Removal**
Now we remove stop words from collected tokens.

```python
import codecs

stop_path = 'stopwords.txt'
stopwords = [normalizer.normalize(x.strip()) for x in codecs.open(stop_path,'r','utf-8').readlines()]
```

```python
tokens_nonstop = [t for t in tokens if t not in stopwords]
title_tokens_nonstop = [t for t in title_tokens if t not in stopwords]
abstract_tokens_nonstop = [t for t in abstract_tokens if t not in stopwords]
paragraphs_tokens_nonstop = [t for t in paragraphs_tokens if t not in stopwords]
tag_tokens_nonstop = [t for t in tag_tokens if t not in stopwords]
```

```python
df = {}

df['title_tokens_nonstop'] = FreqDist(eval('title_tokens_nonstop')).most_common(10) 
df['abstract_tokens_nonstop'] = FreqDist(eval('abstract_tokens_nonstop')).most_common(10) 
df['paragraphs_tokens_nonstop'] = FreqDist(eval('paragraphs_tokens_nonstop')).most_common(10) 
df['tag_tokens_nonstop'] = FreqDist(eval('tag_tokens_nonstop')).most_common(10) 
df['tokens_nonstop'] = FreqDist(eval('tokens_nonstop')).most_common(10) 

freq_analysis_ns = pd.DataFrame(df) 
```

```python
freq_analysis_ns
```

### **Lemmatization**
in order to merge relevant tokens, we use lemmatization.

```python
lemmatizer = Lemmatizer()

title =[lemmatizer.lemmatize(tok) for tok in title_tokens_nonstop] 
abstract = [lemmatizer.lemmatize(tok) for tok in abstract_tokens_nonstop]
paragraphs = [lemmatizer.lemmatize(tok) for tok in paragraphs_tokens_nonstop]
tags = [lemmatizer.lemmatize(tok) for tok in tag_tokens_nonstop]
all_tokens = [lemmatizer.lemmatize(tok) for tok in tokens_nonstop]
```

```python
dflm = {}

dflm['title'] = FreqDist(eval('title')).most_common(10) 
dflm['abstract'] = FreqDist(eval('abstract')).most_common(10) 
dflm['paragraph'] = FreqDist(eval('paragraphs')).most_common(10) 
dflm['tag'] = FreqDist(eval('tags')).most_common(10) 
dflm['all_tokens'] = FreqDist(eval('all_tokens')).most_common(10) 

freq_analysis_lm = pd.DataFrame(dflm) 
```

**Now we can see which words are most used in our data.**

```python
freq_analysis_lm
```

## **Final Task**
Now we can easily extract keywords from the text.

```python
import yake

kw_extractor = yake.KeywordExtractor()
altk = ' '.join(all_tokens)
keywords = kw_extractor.extract_keywords(altk)

for kw in keywords: 
    print(kw)
```
